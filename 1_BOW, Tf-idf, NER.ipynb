```python
# librarys:

import warnings

import spacy,nltk
from spacy import displacy
import pandas as pd

warnings.filterwarnings('ignore')

stopwords = nltk.corpus.stopwords.words('spanish')
nlp_es = spacy.load('es_core_news_lg')
```

#### PART 1.

#### Bag of Words

```python
## To begin with, let's see what the proposed file looks like: hotel.csv
## It is a CSV with two columns, one for text and the other for categories.
## As we are only interested in the text column, we read the file in a dataframe, 
## and we keep the first column in the form of a list.

reviews_hotel = pd.read_csv('/Users/jaime2/Desktop/DATA SCIENCE/5. ANALISIS DE INFORMACION NO ESTRUCTURADA/ENTREGAS/ENTREGA 1/hotel.csv')

print(reviews_hotel.head())

docs = reviews_hotel.iloc[:,0]     # we extract the first column
```

```python
# We obtain the BOW matrix from the vectorizer

from sklearn.feature_extraction.text import CountVectorizer
BOW_vectorizer = CountVectorizer(max_features=20, strip_accents="ascii", lowercase=True)
# BOW_vectorizer = CountVectorizer(max_features=20, strip_accents="ascii", lowercase=True, stop_words = stopwords)

BOW_matrix = BOW_vectorizer.fit_transform(docs)

vocab = BOW_vectorizer.get_feature_names()
pd.DataFrame(BOW_matrix.toarray(), columns=vocab)
```

```python
# And now we do the same for the matrix Tf-idf

from sklearn.feature_extraction.text import TfidfVectorizer
Tfidf_vectorizer = TfidfVectorizer(max_features=20, strip_accents="ascii", lowercase=True)
# Tfidf_vectorizer = TfidfVectorizer(max_features=20, strip_accents="ascii", lowercase=True, stop_words = stopwords)

TF_matrix = Tfidf_vectorizer.fit_transform(docs)

vocab = Tfidf_vectorizer.get_feature_names()
pd.DataFrame(TF_matrix.toarray(), columns=vocab)
```

**Interpretation**.

Typical of a Bag of Words, in the rows we can find each of the elements that make up the corpus, and in the columns all the words that appear in the corpus, the fields being the number of times that word appears in each document. 

One of the traditional problems with Bag-of-Words models is that terms that occur very frequently can detract from the importance of more specific terms that occur less frequently. In fact, two problems occur in BOW:

- Each cell represents how many times a word appears in a document, regardless of how many words that document has.
- Each cell does not store information about how many documents in the total corpus contain that word. 
- The TF-IDF model attempts to solve these drawbacks by introducing a normalization factor.

```python
# load data

espa=open("/Users/jaime2/Desktop/DATA SCIENCE/5. ANALISIS DE INFORMACION NO ESTRUCTURADA/ENTREGAS/ENTREGA 1/esp.txt", "r", encoding="utf-8").read()
print(espa)
```

```python
## We run the analysis pipeline and divide into phrases
docum_analizado = nlp_es(espa)

for sent in docum_analizado.sents:
    print(sent.text)
```

```python
## Print tokens, canonical forms and recognized entities (Not requested in practice)

## POS tagging with Spacy 
spacy_pos_tagged = [(palabra.text, palabra.lemma_, palabra.pos_) for palabra in docum_analizado]
pd.DataFrame(spacy_pos_tagged, columns=['Palabra', 'F. Can√≥nica', 'POS']).T
```

```python
## We print entities

ents_tagged = [(ent.text, ent.label_) for ent in docum_analizado.ents]
pd.DataFrame(ents_tagged, columns=['Entidad', 'Tipo']).T
```

```python
## Entities in the text

for sent in docum_analizado.sents:
    displacy.render(sent, style='ent', jupyter=True)
```
